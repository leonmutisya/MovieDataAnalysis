### Rating Re-Categorization

```python
#create our own function
def rating(rating):
        if rating>=7.0:
            return "Excellent"
        elif rating>=6.0:
            return "Good"
        else:
            return "Average"
```

```python
#lets create new column for this categories
data['Rating_category']=data['Rating'].apply(rating)
```

```python
data.columns
```

```python
data['Genre']
```

```python
#Split the genres using commas and make an empty list
list1=[]
for value in data['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
#we going to add this 2list to 1list
one_d=[]
for item in list1:
        for item1 in item:
            one_d.append(item1)
```

### Finding Unique Values in Genres

```python
#Finding unique values
uni_list=[]
for item in one_d:
        if item not in uni_list:
            uni_list.append(item)
```

```python
print (uni_list)
```

### Visualizing the Genre Count

```python
#lets visualize it
# Convert the flattened list into a DataFrame
data_one_d = pd.DataFrame(one_d, columns=['Genre'])

# Count the occurrences of each category
genre_counts = data_one_d['Genre'].value_counts().reset_index()

# Rename the columns for clarity
genre_counts.columns = ['Genre', 'Count']

# Create a bar plot to visualize the number of occurrences per category
sns.barplot(x='Count', y='Genre', data=genre_counts, orient='h')
plt.title("Number of films Per Genre")
plt.xlabel("Number of Occurrences")
plt.ylabel("Genre")
plt.show()
```
